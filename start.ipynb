```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup



browser = webdriver.Chrome(executable_path = 'C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe')

wait = WebDriverWait(browser, 10)
def next_page():
    try:
        browser.get('https://www.startupranking.com/top/united-states/')
        submit = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR,'body > div.wrapper.container > section > button.btn-blue.load_previous_ranks'))
        )
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        submit.click()
    except TimeoutException:
        return next_page()
def next_page2():
    try:
        submit = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR,'body > div.wrapper.container > section > button:nth-child(8)'))
        )
        #page =  browser.find_element_by_css_selector('body > div.wrapper.container > section > button:nth-child(8)')
        #browser.execute_script("arguments[0].scrollIntoView(false);", page)
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(3)
        submit.click()
    except TimeoutException:
        return next_page2()
def get_company_name(name,sr_score,description,rank,country):
    html = browser.page_source
    soup = BeautifulSoup(html,'html')
    for n in soup.find('tbody',class_='ranks').find_all('div',class_='name'):
        name.append(n.text)
    for s in soup.find('tbody',class_="ranks").find_all('td',class_= 'tright sr-score'):
        sr_score.append(s.text.strip())
    for d in soup.find('tbody',class_="ranks").find_all('td',class_= 'tleft description'):
        description.append(d.text.strip())
    for r in soup.find('tbody',class_="ranks").find_all('div', class_= 'country-rank'):
        rank.append(r.text)
    for c in soup.find('tbody',class_="ranks").find_all('img'):
        country.append(c.get('title'))
        
def click_company(company_name):
    try:
        company_name = company_name.lower()
        submit = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR,'#startup_'+company_name+' > td:nth-child(2) > div.name > a'))
        )
        time.sleep(1)
        submit.click()
    except TimeoutException:
        return click_company(company_name)()

    
def get_company_location(name,location):
    html = browser.page_source
    soup = BeautifulSoup(html,'html')
    name_location.append(name)
    location.append(soup.find('div',class_='su-logo').find('ul',class_='su-ranks').find('li',class_ = 'su-ranks').text.strip())

def get_company_location2(name,location,found_time):
    results = browser.find_elements_by_css_selector("body > div.wrapper.container.group > section.general-info.group > div.su-logo > ul:nth-child(2) > li.su-state > a")
    
    if results is None:
        location.append("Unknown")
        name_location.append(name)
    else:
        for result in results:
            link = result.find_element_by_xpath("/html/body/div[5]/section[1]/div[1]/ul[1]/li[3]/a")
            location.append(link.text.strip())
            name_location.append(name)
    
    results2 = browser.find_elements_by_css_selector("body > div.wrapper.container.group > section.general-info.group > div.su-info > p.su-loc")
    if results2 is None:
        found_time.append("Unknown")
    else:
        for result in results2:
            link2 = result.find_element_by_xpath("/html/body/div[5]/section[1]/div[3]/p[3]")
            found_time.append(link2.text.strip())


def get_company_founder(name,nameofcompany,founder,role,period):
    html = browser.page_source
    soup = BeautifulSoup(html,'html')
    testfoundername = soup.find('tbody',class_="ranks").find('div',class_="name")
    if testfoundername is None:
        founder.append('Unkonwn')
        nameofcompany.append(name)
        role.append("Unkonwn")
        period.append("Unkonwn")
    else:
        results = browser.find_elements_by_css_selector("body > div.wrapper.container.group > section:nth-child(19) > table > tbody > tr:nth-child(1) > td.tleft > div.name > a")
        for result in results:
            found = result.find_element_by_xpath("/html/body/div[5]/section[8]/table/tbody/tr[1]/td[1]/div[2]/a")
            founder.append(found.text.strip())
            nameofcompany.append(name)
        results2 = browser.find_elements_by_css_selector("body > div.wrapper.container.group > section:nth-child(19) > table > tbody > tr:nth-child(1) > td:nth-child(2)")
        for result in results2:
            roleofhim = result.find_element_by_xpath("/html/body/div[5]/section[8]/table/tbody/tr[1]/td[2]")
            role.append(roleofhim.text.strip())
        results3 = browser.find_elements_by_css_selector("body > div.wrapper.container.group > section:nth-child(19) > table > tbody > tr:nth-child(1) > td:nth-child(3)")
        for result in results3:
            pe = result.find_element_by_xpath("/html/body/div[5]/section[8]/table/tbody/tr[1]/td[3]")
            period.append(pe.text.strip())
        
def get_company_message(nameofcompany,founder,role,period,location,found_time):
    html = browser.page_source
    soup = BeautifulSoup(html,'html')
    list1=[]
    for n in soup.find('tbody',class_='ranks').find_all('div',class_='name'):
        list1.append(n.text)
    for name in list1:
        name = name.lower()
        name = name.replace('.','-')
        name = name.replace(',','-')
        name = name.replace(' ','-')
        name = name.replace('. ','')
        try:
            page =  browser.find_element_by_css_selector('#startup_'+name+' > td:nth-child(2) > div.name > a')
            browser.execute_script("arguments[0].scrollIntoView(false);", page)
        except:
            continue
        click_company(name)
        get_company_location2(name,location,found_time)
        get_company_founder(name,nameofcompany,founder,role,period)
        browser.back()

def main():
    next_page()
    get_company_message(nameofcompany,founder,role,period,location,found_time)
    for i in range(60):
        next_page2()
        get_company_message(nameofcompany,founder,role,period,location,found_time)
#    for i in range(10):
#        next_page2()
#        get_company_name(name,sr_score,description,rank,country)

nameofcompany =[]
founder=[]
role = []
period=[]
location=[]
name_location=[]
found_time = []

if __name__ == "__main__":
    main()

arr2 = np.array([nameofcompany,founder,role,period])

founder_info = pd.DataFrame(arr2.T,columns = ['Name','Founder','Role','Period'])   

arr3 = np.array([name_location,location,found_time])

location_info= pd.DataFrame(arr3.T,columns = ['Name','Location','Founded time'])   


```
